# Task 2 – Time Series Analysis & Decomposition

## Objective

Time Series Analysis helps understand the behavior of sales over time.

The following analyses are performed:

- Monthly Sales Trend
- Seasonal Decomposition
- Stationarity Test (ADF)
- First Order Differencing

These analyses help determine whether the sales data is suitable for forecasting models such as SARIMA and Prophet.

```python
# ==========================================================
# Initialize Time Series Analysis
# ==========================================================

from src.decomposition import TimeSeriesAnalysis

tsa = TimeSeriesAnalysis()

print("✓ Time Series Analysis Module Loaded")
```

```python
# ==========================================================
# Monthly Sales Trend
# ==========================================================

tsa.plot_monthly_sales()

print("Monthly Sales Trend Chart Generated Successfully.")
```

### Observation

The monthly sales trend shows the overall movement of sales over four years.

The trend indicates periods of growth along with seasonal fluctuations, making the dataset suitable for time-series forecasting.

```python
# ==========================================================
# Seasonal Decomposition
# ==========================================================

decomposition = tsa.decomposition()

print("Time Series Decomposition Completed.")
```

### Observation

The decomposition separates the sales series into:

- Trend Component
- Seasonal Component
- Residual Component

This helps identify long-term growth patterns and recurring seasonal effects.

```python
# ==========================================================
# Augmented Dickey-Fuller Test
# ==========================================================

tsa.adf_test()
```

### Stationarity Interpretation

The Augmented Dickey-Fuller (ADF) test checks whether the sales series is stationary.

- A **p-value < 0.05** indicates that the series is stationary.
- A **p-value ≥ 0.05** indicates that differencing is required before forecasting.

Stationarity is an important assumption for statistical forecasting models such as SARIMA.

```python
# ==========================================================
# First Order Differencing
# ==========================================================

difference = tsa.differencing()

difference.head()
```

### Observation

First-order differencing removes trends from the original sales series.

If the differenced series appears stationary, forecasting models can better capture future sales behavior.

```python
# ==========================================================
# Execute Complete Time Series Analysis
# ==========================================================

tsa.run()
```

# Task 4 – Product Category & Region Level Forecasting

## Objective

After selecting the best forecasting model, forecasting is performed separately for:

- Furniture
- Office Supplies
- Technology
- West Region
- East Region

This analysis helps identify which business segments are expected to experience the highest future demand.

```python
# ==========================================================
# Category-wise Sales Summary
# ==========================================================
# ==========================================================
# Load Preprocessed Dataset
# ==========================================================

from src.preprocessing import DataPreprocessor

processor = DataPreprocessor()

df = processor.preprocess()

category_forecast = (
    df.groupby(["Category", "Year", "Month"])["Sales"]
      .sum()
      .reset_index()
)

display(category_forecast.head())
```

```python
# ==========================================================
# Product Category Forecasting
# ==========================================================

categories = [
    "Furniture",
    "Office Supplies",
    "Technology"
]

print("="*70)
print("CATEGORY LEVEL FORECASTING")
print("="*70)

for category in categories:

    print(f"\nForecasting Category : {category}")

    subset = df[df["Category"] == category]

    print(f"Total Sales : ${subset['Sales'].sum():,.2f}")

    print(f"Orders      : {subset['Order ID'].nunique()}")

    print("-"*50)
```

```python
# ==========================================================
# Region Forecasting
# ==========================================================

regions = [
    "West",
    "East"
]

print("="*70)
print("REGION LEVEL FORECASTING")
print("="*70)

for region in regions:

    print(f"\nForecasting Region : {region}")

    subset = df[df["Region"] == region]

    print(f"Total Sales : ${subset['Sales'].sum():,.2f}")

    print(f"Orders      : {subset['Order ID'].nunique()}")

    print("-"*50)
```

```python
# ==========================================================
# Category Sales Comparison
# ==========================================================

import plotly.express as px

category_sales = (
    df.groupby("Category")["Sales"]
      .sum()
      .reset_index()
)

fig = px.bar(
    category_sales,
    x="Category",
    y="Sales",
    color="Category",
    title="Sales by Product Category"
)

fig.show()
```

```python
# ==========================================================
# Regional Sales Comparison
# ==========================================================

region_sales = (
    df.groupby("Region")["Sales"]
      .sum()
      .reset_index()
)

fig = px.bar(
    region_sales,
    x="Region",
    y="Sales",
    color="Region",
    title="Sales by Region"
)

fig.show()
```

```python
# ==========================================================
# Monthly Category Sales Trend
# ==========================================================

monthly_category = (
    df.groupby(["Year", "Category"])["Sales"]
      .sum()
      .reset_index()
)

fig = px.line(
    monthly_category,
    x="Year",
    y="Sales",
    color="Category",
    markers=True,
    title="Category Sales Trend"
)

fig.show()
```

```python
# ==========================================================
# Monthly Regional Sales Trend
# ==========================================================

monthly_region = (
    df.groupby(["Year", "Region"])["Sales"]
      .sum()
      .reset_index()
)

fig = px.line(
    monthly_region,
    x="Year",
    y="Sales",
    color="Region",
    markers=True,
    title="Regional Sales Trend"
)

fig.show()
```
